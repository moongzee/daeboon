# Softmax Score

```python
Image('KakaoTalk_20210303_113836786_05.png')
```

```python
Image('KakaoTalk_20210303_113836786_06.png')
```

```python
import torch
import torch.nn as nn
import numpy as np

def softmax(x):
    return np.exp(x)/np.sum(np.exp(x),axis=0)

x=np.array([2.0,1.0,0.1])
outputs=softmax(x)
print('softmax numpy:',outputs)

x=torch.tensor([2.0,1.0,0.1])
outputss=torch.softmax(x,dim=0)
print('torch softmax numpy:',outputss)
```

# Cross-Entropy loss

```python
Image('KakaoTalk_20210303_113836786_07.png')
```

```python
import torch
import torch.nn as nn
import numpy as np

def cross_entropy(actual,predicted):
    loss=(np.sum(actual*np.log(predicted)))
    return loss
          
Y=np.array([1,0,0])

Y_pred_good=np.array([0.7,0.2,0.1])
Y_pred_bad=np.array([0.1,0.3,0.6])
loss1=cross_entropy(Y,Y_pred_good)
loss2=cross_entropy(Y,Y_pred_bad)

print(f'Loss1:{loss1:.4f}')
print(f'Loss2:{loss2:.4f}')
```
