```python
a=['CT','cartilage','implant','mandible','maxilla','skin','skull']
b=[sum(CT['CT']),sum(CT['cartilage']),sum(CT['implant']),sum(CT['mandible']),sum(CT['maxilla']),sum(CT['skin']),sum(CT['skull'])]
keyword_df=pd.DataFrame(b,a)
keyword_df
```

```python
a=CT.loc[:,'Spacing_X':'Dimension_Z']

sns.clustermap(a.corr(),figsize=(10,10), annot=True, cmap='RdYlBu_r', vmin=-1, vmax=1)
```

```python
a=CT.loc[:,'Spacing_X':'keywords_count']

plt.figure(figsize=(12,12))
sns.heatmap(a.corr(), annot=True, linewidths=.5, cmap='Greens', vmin=0, vmax=1, cbar_kws={'shrink':.5})
```

```python
# Spacing_Z 와 Dimension_Z와의 상관관계 // High negative corr
import numpy as np
N=CT['idx'].count()
x=CT['Dimension_Z']
y=CT['Spacing_Z']
colors=np.random.rand(N)
area=(30*np.random.rand(N))**2

plt.scatter(x,y,s=area,c=colors,alpha=0.5)
plt.show()
```

```python
#Spacing_Z 와 Dimension_Z lmplot
sns.lmplot(data=CT, x='Spacing_Z',y='Dimension_Z')
```

```python
#Spacing_Z 와 Dimension_Z jointplot
sns.set({'figure.figsize':(7,7)})
g=sns.jointplot('Spacing_Z','Dimension_Z',data=CT,kind='kde',space=0)
plt.tight_layout()
plt.show()
```

```python
#Spacing 좌표 3D scatter 그리기

from mpl_toolkits.mplot3d import Axes3D

N=CT['idx'].count()
X=CT['Spacing_X']
Y=CT['Spacing_Y']
Z=CT['Spacing_Z']
color=20*np.random.rand(N)
plt.rcParams['figure.figsize']=(10,10)
fig=plt.figure()

ax=fig.add_subplot(111,projection='3d')

ax.scatter(X,Y,Z,c=color,marker='o',s=15,cmap='Greens')
```
